```python
from azureml import Workspace

ws = Workspace()
ds = ws.datasets['data.csv']
dataset = ds.to_dataframe()
```

```python
dataset
```

```python
cols = dataset.columns.drop('Food Name')
```

```python
import pandas as pd
dataset[cols] = dataset[cols].apply(pd.to_numeric, errors='coerce')
```

```python
dataset
```

```python
dataset = dataset.fillna(0)
```

```python
X = dataset.drop('Food Name',1)
```

```python
X
```

```python
Y = dataset["Food Name"]
```

```python
Y[:5]
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
```

```python
clf = clf.fit(X,Y)
```

```python
X[:1]
```

```python
clf.predict(X[:1])
```

```python
Y[:1]
```

```python
# save the model to disk
import pickle
filename = 'decisionTree_model.sav'
pickle.dump(clf, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.predict(X[:5])
print(result)
```
