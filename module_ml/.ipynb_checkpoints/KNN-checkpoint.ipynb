```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
dataset = pd.read_csv('data.csv')
dataset.head()
```

```python
dataset = dataset.fillna(0)
dataset.describe()
```

```python
dataset.to_csv('tt.csv')
```

```python
cols = dataset.columns.drop('Food Name')
```

```python
dataset[cols] = dataset[cols].apply(pd.to_numeric, errors='coerce', downcast='integer')
```

```python
dataset = dataset.fillna(0)
```

```python
X = dataset.drop('Food Name',1)
X.head(5)
```

```python
sns.pairplot(X);
```

```python
X.describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
```

```python
#sns.pairplot(X_pca);
```

```python
Y = dataset["Food Name"]
```

```python
#X = X/X.mean()
#normalizer = preprocessing.Normalizer().fit(X)  
#X = normalizer.transform(X)
#print(X)

#scaler = StandardScaler()
#scaler.fit_transform(X)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-3,3))
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.10, random_state=42)
```

```python

data = [[10, 20, 0.1, 1.0,3.0, 0.0, 0.2, 0.0, 1.0, 50.0, 1.5]]
#data = data /np.mean(data)

```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=10, metric='cosine')
neigh.fit(X_train, Y_train)

```

```python
Y_pred = neigh.predict(X_test)

# compare actual response values (y_test) with predicted response values (y_pred)
print(metrics.accuracy_score(Y_test, Y_pred))
```

```python
print(data)
print(neigh.kneighbors(data))

print(Y.iloc[1362])
print(Y.iloc[63])
print(Y.iloc[590])
print(Y.iloc[61])
print(Y.iloc[273])
```
