```python
import pandas as pd
import pickle
```

```python
dataset = pd.read_csv('data.csv')
dataset.head()
```

```python
cols = dataset.columns.drop('Food Name')
```

```python
dataset[cols] = dataset[cols].apply(pd.to_numeric, errors='coerce')
```

```python
dataset = dataset.fillna(0)
```

```python
X = dataset.drop('Food Name',1)
X
```

```python
Y = dataset["Food Name"]
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
```

```python
clf = clf.fit(X,Y)
```

```python
clf.predict(X[:1])
```

```python

filename = 'decisionTree_model.sav'
pickle.dump(clf, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.predict(X[:5])
print(result)
```
